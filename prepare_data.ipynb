```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

num_features = 300    # Word vector dimensionality
min_word_count = 1   # Minimum word count
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size
downsampling = 1e-3   # Downsample setting for frequent words

print ("Training model...")
model = word2vec.Word2Vec(data_final_1, workers=num_workers, size=num_features, min_count = min_word_count,
                            window = context, sample = downsampling, iter=4 , sg =1 )
model.init_sims(replace=True)
# model.save('./model/word2vec_must_dice')
model.save_word2vec_format('word2vec_must_dice.bin', binary=True)

```
